````python
def extract_text_from_file(file_path):
    """
    PDF ë˜ëŠ” Word(.docx) íŒŒì¼ì—ì„œ í…ìŠ¤íŠ¸ë¥¼ ì¶”ì¶œí•˜ê³ , ì¤„ë°”ê¿ˆ ë° ë„ì–´ì“°ê¸°ë¥¼ ì •ì œí•˜ëŠ” í•¨ìˆ˜

    Args:
        file_path (str): íŒŒì¼ ê²½ë¡œ (PDF ë˜ëŠ” DOCX)

    Returns:
        str: ì •ì œëœ í…ìŠ¤íŠ¸
    """
    ext = os.path.splitext(file_path)[-1].lower()
    text = ""

    if ext == ".pdf":
        try:
            reader = PdfReader(file_path)
            for page in reader.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + "\n"
        except Exception as e:
            print(f"[PDF ì½ê¸° ì˜¤ë¥˜] {e}")

    elif ext == ".docx":
        try:
            doc = Document(file_path)
            for para in doc.paragraphs:
                text += para.text.strip() + "\n"
        except Exception as e:
            print(f"[DOCX ì½ê¸° ì˜¤ë¥˜] {e}")

    else:
        raise ValueError("ì§€ì›í•˜ì§€ ì•ŠëŠ” íŒŒì¼ í˜•ì‹ì…ë‹ˆë‹¤. PDF ë˜ëŠ” DOCXë§Œ ê°€ëŠ¥í•©ë‹ˆë‹¤.")

    # í›„ì²˜ë¦¬: ì¤„ë°”ê¿ˆ ì •ì œ
    # 1. ë¬¸ì¥ ì¤‘ê°„ì˜ ì¤„ë°”ê¿ˆ(\n)ì€ ë„ì–´ì“°ê¸°ë¡œ ì¹˜í™˜
    text = re.sub(r"(?<!\n)\n(?!\n)", " ", text)

    # 2. ë‘ ê°œ ì´ìƒì˜ ì—°ì† ì¤„ë°”ê¿ˆì€ ë¬¸ë‹¨ êµ¬ë¶„ìœ¼ë¡œ ë³´ê³  ìœ ì§€ (í•˜ë‚˜ì˜ \nìœ¼ë¡œ)
    text = re.sub(r"\n{2,}", "\n", text)

    # 3. ì—°ì† ê³µë°± ì •ë¦¬
    text = re.sub(r"[ \t]{2,}", " ", text)

    return text.strip()
````

````python
def search_perplexity(query, api_key):
    url = "https://api.perplexity.ai/chat/completions"
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    payload = {
        "model": "sonar-pro",
        "messages": [{"role": "user", "content": query}],
        "return_citations": True  # ì¤‘ìš”: citation í¬í•¨í•˜ë„ë¡ ì„¤ì •
    }

    response = requests.post(url, headers=headers, json=payload)
    if response.status_code == 200:
        try:
            data = response.json()
            message = data.get("choices", [{}])[0].get("message", {})
            content = message.get("content", "")

            # âœ… ì‹¤ì œ citation URLì€ 'search_results' í•„ë“œì— ìˆìŒ
            search_results = data.get("search_results", [])
            urls = []
            for result in search_results:
                url = result.get("url")
                if url:
                    urls.append(url)

            return content.strip(), urls
        except Exception as e:
            print("Perplexity Parsing Error:", e)
            return "", []
    else:
        print("Perplexity API Error:", response.status_code, response.text)
        return "", []


def search_serpapi_with_url(query, serpapi_key):
    """
    SerpAPIë¥¼ ì‚¬ìš©í•˜ì—¬ Google ê²€ìƒ‰ ê²°ê³¼ ìš”ì•½ê³¼ í•¨ê»˜ URL ë°˜í™˜
    """
    url = "https://serpapi.com/search"
    params = {
        "engine": "google",
        "q": query,
        "api_key": serpapi_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        results = response.json()
        snippets = []
        urls = []
        for r in results.get("organic_results", [])[:3]:
            snippet = r.get("snippet", "")
            link = r.get("link", "")
            if snippet and link:
                snippets.append(snippet)
                urls.append(link)
        return "\n".join(snippets), urls
    else:
        print("SerpAPI Error:", response.status_code, response.text)
        return "", []

def extract_relevant_summary_from_content(slide_description, elements, research_result):
    """
    ë¦¬ì„œì¹˜ ê²°ê³¼ ì¤‘ slide_description ë° elements ë‚´ìš©ì— ë¶€í•©í•˜ëŠ” í•µì‹¬ ì •ë³´ë§Œ ìš”ì•½ ì¶”ì¶œ
    ë‹¨, ìš”ì•½ì€ ì§€ë‚˜ì¹˜ê²Œ ì¶•ì•½ë˜ì§€ ì•Šë„ë¡ 3~5ë¬¸ì¥ ë‚´ì™¸ë¡œ ê°„ê²°íˆ ì‘ì„±
    """
    prompt = (
        f"ë‹¹ì‹ ì€ ì œì•ˆì„œ ì‘ì„± ì§€ì› AIì…ë‹ˆë‹¤.\n"
        f"ë‹¤ìŒì€ ìŠ¬ë¼ì´ë“œì˜ ëª©ì ê³¼ í¬í•¨ëœ ìš”ì†Œë“¤, ê·¸ë¦¬ê³  ë¦¬ì„œì¹˜ ê²°ê³¼ì…ë‹ˆë‹¤.\n\n"
        f"[ìŠ¬ë¼ì´ë“œ ëª©ì ]\n{slide_description}\n\n"
        f"[ìŠ¬ë¼ì´ë“œ ìš”ì†Œ]\n{elements}\n\n"
        f"[ë¦¬ì„œì¹˜ ê²°ê³¼]\n{research_result}\n\n"
        f"ë¦¬ì„œì¹˜ ê²°ê³¼ì¤‘ ì¶œì²˜ê°€ blogì™€ ê°™ì´ ì‹ ë¢°í•˜ê¸° í˜ë“  ë°ì´í„°ëŠ” ì œì™¸í•´ì£¼ì„¸ìš”.\n\n"
        f"ë¦¬ì„œì¹˜ ê²°ê³¼ì™€ elementsë‚´ ë“¤ì–´ê°€ëŠ” ë‚´ìš©, urlì„ ê¼­ ë§¤ì¹­í•´ì£¼ì„¸ìš”. ex). [ROI 20%, AI ê¸°ë°˜ ìë™í™” ëª¨ë“ˆì„ í†µí•´ ì—…ë¬´ íš¨ìœ¨ì„±ì„ í¬ê²Œ ë†’ì¸ ë°” ìˆìœ¼ë©°, ì´ëŠ” ROI 20%ì™€ TCO ì ˆê° 15%ë¥¼ ë‹¬ì„±í•˜ëŠ” ë° ê¸°ì—¬í•©ë‹ˆë‹¤., https://botpress.com/ko/blog/top-artificial-intelligence-trends\n\n"
        f"ë¦¬ì„œì¹˜ ê²°ê³¼ì¤‘ ì¥í‘œì— ë“¤ì–´ê°€ì•¼í•˜ëŠ” ìˆ˜ì¹˜ëŠ” ëª¨ë‘ research_resultì—ì„œ ë‚˜ì˜¨ ìˆ˜ì¹˜ê°€ ê¼­ í¬í•¨ì´ ë ìˆ˜ ìˆë„ë¡ í•´ì£¼ì„¸ìš” ex). ì„±ì¥ë¥ , íš¨ìœ¨ì„± í–¥ìƒë¥ , ì‹œì¥ ê·œëª¨ ë“±\n\n"
        f"ìœ„ ìŠ¬ë¼ì´ë“œ ëª©ì ê³¼ elementsì™€ ìŠ¬ë¼ì´ë“œ ëª©ì ì— ë¶€í•©í•˜ëŠ” í•µì‹¬ ì •ë³´ë§Œ ìš”ì•½í•˜ì„¸ìš”. ë‹¨, elementsë‚´ ìˆ˜ì¹˜ ë°ì´í„° ë° í•„ìš”í•œ ë°ì´í„°ëŠ” ê¼­ í¬í•¨í•˜ì„¸ìš”.\nn"
        f"ë¶ˆí•„ìš”í•œ ì¼ë°˜ë¡ , ë„ì…ë¶€, ì¤‘ë³µ í‘œí˜„ì€ ì œê±°í•˜ì„¸ìš”. ìš”ì•½ì€ 3~5ë¬¸ì¥ìœ¼ë¡œ ìœ ì§€í•˜ì„¸ìš”."
    )
    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()


def check_relevance(description, elements, research_result):
    """
    ë¦¬ì„œì¹˜ ê²°ê³¼ê°€ slide_descriptionê³¼ ë…¼ë¦¬ì ìœ¼ë¡œ ë¶€í•©í•˜ëŠ”ì§€ í™•ì¸
    """
    prompt = (
        "ë‹¹ì‹ ì€ ì „ëµ ì»¨ì„¤íŒ… ì œì•ˆì„œ ê²€í†  ì „ë¬¸ê°€ì…ë‹ˆë‹¤.\n\n"
        "ë‹¤ìŒì€ íŠ¹ì • ìŠ¬ë¼ì´ë“œì˜ ëª©ì  ì„¤ëª…ê³¼, ìµœì‹  ë¦¬ì„œì¹˜ ê²°ê³¼ì…ë‹ˆë‹¤. ìŠ¬ë¼ì´ë“œì˜ elementsëŠ” ê¸°ì¡´ GPT ëª¨ë¸ ê¸°ë°˜ ì´ˆì•ˆì´ë©°, ì •ë³´ê°€ ì˜¤ë˜ë˜ì—ˆê±°ë‚˜ ëœ ìœ ì˜ë¯¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.\n"
        "ë°˜ë©´ ì•„ë˜ ë¦¬ì„œì¹˜ ê²°ê³¼ëŠ” ìµœì‹ ì´ê³ , ì‹ ë¢° ê°€ëŠ¥í•œ ì •ë³´ì›ì—ì„œ ê°€ì ¸ì™”ìŠµë‹ˆë‹¤.\n\n"
        "[ìŠ¬ë¼ì´ë“œ ëª©ì ]\n"
        f"{description}\n\n"
        "[ìŠ¬ë¼ì´ë“œ ë‚´ìš©]\n"
        f"{elements}\n\n"
        "[ë¦¬ì„œì¹˜ ê²°ê³¼]\n"
        f"{research_result}\n\n"
        "ğŸ“Œ ê²€í†  ê¸°ì¤€:\n"
        "- ë¦¬ì„œì¹˜ ê²°ê³¼ê°€ ìŠ¬ë¼ì´ë“œì˜ ëª©ì ì— ì‹¤ì§ˆì ìœ¼ë¡œ ê¸°ì—¬í•˜ê±°ë‚˜, ìŠ¬ë¼ì´ë“œì— í¬í•¨ë˜ë©´ ì •ë³´ì˜ í’ˆì§ˆì´ë‚˜ ì„¤ë“ë ¥ì´ í–¥ìƒë˜ëŠ” ê²½ìš°, 'ë„¤'ë¼ê³  ë‹µí•˜ì‹­ì‹œì˜¤.\n"
        "- ë‹¨ìˆœíˆ ë°°ê²½ ì„¤ëª… ìˆ˜ì¤€ì´ê±°ë‚˜ ëª©ì ê³¼ ë™ë–¨ì–´ì§„ ë‚´ìš©ì´ë©´ 'ì•„ë‹ˆì˜¤'ë¡œ ë‹µí•˜ì‹­ì‹œì˜¤.\n"
        "- ì§ˆë¬¸ê³¼ ê´€ë ¨ ì—†ëŠ” ë°ì´í„°ì˜ ê²½ìš° ê³¼ê°í•˜ê²Œ ì œì™¸í•´ì£¼ì„¸ìš”. ex). AI ëª¨ë¸ í•™ìŠµ ë° í‰ê°€ ìë™í™”ì˜ í˜„ì¬ ì‹œì¥ ê·œëª¨ì™€ ê´€ë ¨ ì—†ëŠ” ê²°ê³¼  'ì§ˆë¬¸: AI ëª¨ë¸ í•™ìŠµ ë° í‰ê°€ ìë™í™”ì˜ í˜„ì¬ ì‹œì¥ ê·œëª¨ì™€ í–¥í›„ 5ë…„ê°„ì˜ ì˜ˆìƒ ì„±ì¥ë¥ ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”?' ë‹µë³€: 'AI êµìœ¡ ë°ì´í„° ì„¸íŠ¸ ì‹œì¥ì€ 2024ë…„ 292ì–µ ë‹¬ëŸ¬ì—ì„œ 2032ë…„ 1,740ì–µ ë‹¬ëŸ¬ë¡œ ì„±ì¥í•  ì „ë§ì´ë©°, ì„œë¹„ìŠ¤í˜• AI ì‹œì¥ì€ 2024ë…„ 127ì–µ ë‹¬ëŸ¬ì—ì„œ ì—°í‰ê·  30.6% ì„±ì¥í•  ê²ƒìœ¼ë¡œ ì˜ˆìƒë©ë‹ˆë‹¤.'\n"
        "- ìˆ˜ì¹˜ ì°¨ì´ë‚˜ ì •í™•ë„ë³´ë‹¤ ìŠ¬ë¼ì´ë“œ ì‘ì„±ì— ë„ì›€ì´ ë˜ëŠ” ì‹¤ìš©ì  ì •ë³´ì¸ì§€ë¥¼ ì¤‘ì ì ìœ¼ë¡œ íŒë‹¨í•˜ì„¸ìš”.\n\n"
        "ë‹µë³€ì€ ë°˜ë“œì‹œ 'ë„¤' ë˜ëŠ” 'ì•„ë‹ˆì˜¤'ë¡œ ì‹œì‘í•˜ê³ , 1~2ë¬¸ì¥ìœ¼ë¡œ ì´ìœ ë¥¼ ì„¤ëª…í•´ì£¼ì„¸ìš”.\n"
        "ì˜ˆì‹œ: ë„¤. ì´ ë°ì´í„°ëŠ” ì‹œì¥ ì„±ì¥ë¥ ê³¼ ê´€ë ¨ëœ ì •ë³´ë¡œ, í•´ë‹¹ ìŠ¬ë¼ì´ë“œì˜ ì„¤ë“ë ¥ì„ ë†’ì´ëŠ” ë° ìœ ìš©í•©ë‹ˆë‹¤."
    )

    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1
    )
    answer = response.choices[0].message.content.strip()
    print(f"\n[GPT íŒë‹¨ ê²°ê³¼]\n{answer}\n")
    return answer.startswith("ë„¤")


def regenerate_query(original_query, previous_answer):
    """
    ê²€ìƒ‰ ê²°ê³¼ê°€ ë¶€ì í•©í•œ ê²½ìš°, ì§ˆë¬¸ì„ ë” ëª…í™•í•˜ê²Œ ë§Œë“¤ì–´ì£¼ëŠ” ë¡œì§
    """
    prompt = (
        f"ë‹¤ìŒì€ ì •ë³´ ê²€ìƒ‰ì„ ìœ„í•œ ì›ë˜ ì§ˆë¬¸ê³¼ ê²€ìƒ‰ ê²°ê³¼ì…ë‹ˆë‹¤.\n\n"
        f"[ì§ˆë¬¸]\n{original_query}\n\n"
        f"[ê²€ìƒ‰ ê²°ê³¼]\n{previous_answer}\n\n"
        "ìœ„ ê²°ê³¼ê°€ ë¶€ì •í™•í•˜ê±°ë‚˜ ë¶€ì¡±í•œ ê²½ìš°, ì§ˆë¬¸ì„ ë” êµ¬ì²´ì ì´ê³  ë‹µë³€ì„ ìœ ë„í•  ìˆ˜ ìˆë„ë¡ ì¬ì‘ì„±í•˜ì„¸ìš”. "
        "ìƒˆ ì§ˆë¬¸ì€ 1ë¬¸ì¥ìœ¼ë¡œ ê°„ê²°í•˜ê²Œ ì‘ì„±í•´ ì£¼ì„¸ìš”."
    )
    response = openai.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()

import json

def smart_research_fill(slides_dict, search_mode="perplexity", max_retry=2):
    """
    ìŠ¬ë¼ì´ë“œì— ëŒ€í•´ ìŠ¤ë§ˆíŠ¸ ë¦¬ì„œì¹˜ë¥¼ ìˆ˜í–‰í•˜ê³ , ê´€ë ¨ì„± íŒë‹¨ ê²°ê³¼ë¥¼ ìŠ¬ë¼ì´ë“œ ì œëª©ê³¼ ì§ˆë¬¸ê³¼ í•¨ê»˜ ì¶œë ¥í•¨.
    Perplexity ì‚¬ìš© ì‹œ citation URLë„ í•¨ê»˜ ì €ì¥ë¨.
    """
    completed_results = {}

    for slide_id, slide in slides_dict.items():
        desc = slide.get("description", "")
        if not desc or not isinstance(desc, str):
            print(f"[ê²½ê³ ] descriptionì´ ë¹„ì–´ìˆê±°ë‚˜ ë¬¸ìì—´ì´ ì•„ë‹˜: slide_id={slide_id}")
            continue

        needs_list = slide.get("needs_research", [])
        if not isinstance(needs_list, list) or not needs_list:
            print(f"[ìŠ¤í‚µ] needs_research ì—†ìŒ ë˜ëŠ” ë¹ˆ ë¦¬ìŠ¤íŠ¸ - slide_id={slide_id}")
            continue

        slide_title = slide.get("slide_title", slide_id)
        slide_elements = slide.get("elements", {})

        # JSON ì•ˆì „í•˜ê²Œ ë³€í™˜
        try:
            elements_str = json.dumps(slide_elements, ensure_ascii=False, indent=2)
        except Exception as e:
            print(f"[ê³ ë„í™” ì‹¤íŒ¨] {slide_id}: elements ë³€í™˜ ì‹¤íŒ¨ - {e}")
            continue

        slide_results = {}
        for need in needs_list:
            if not isinstance(need, str):
                continue
            query = need
            attempt = 0

            while attempt < max_retry:
                # ğŸ” ê²€ìƒ‰ ìˆ˜í–‰
                if search_mode == "perplexity":
                    content, urls = search_perplexity(query, perplexity_api_key)
                elif search_mode == "serpapi":
                    content, urls = search_serpapi_with_url(query, serp_api_key)
                else:
                    raise ValueError("search_modeëŠ” 'perplexity' ë˜ëŠ” 'serpapi'ë§Œ ì§€ì›ë©ë‹ˆë‹¤.")

                # â›” ê²€ìƒ‰ ì‹¤íŒ¨ ì²˜ë¦¬
                if not content:
                    print(f"[ì‹¤íŒ¨] ê²€ìƒ‰ ê²°ê³¼ ì—†ìŒ - query: {query}")
                    break

                # âœ… íŒë‹¨ ë¡œê·¸ ì¶œë ¥
                print(f"\n===== [ìŠ¬ë¼ì´ë“œ: {slide_title}] ì§ˆë¬¸: \"{query}\" ì— ëŒ€í•œ íŒë‹¨ =====")
                is_relevant = check_relevance(desc, slide_elements, content)

                if is_relevant:
                    try:
                        summary = extract_relevant_summary_from_content(
                            slide_description=desc,
                            elements=elements_str,  # ë¬¸ìì—´ë¡œ ì „ë‹¬
                            research_result=content
                        )
                    except Exception as e:
                        print(f"[ìš”ì•½ ì‹¤íŒ¨] {slide_id} - {query}: {e}")
                        break

                    slide_results[need] = {
                        "content": summary,
                        "urls": urls
                    }
                    break
                else:
                    query = regenerate_query(query, content)
                    attempt += 1
                    time.sleep(1)

        if slide_results:
            completed_results[slide_id] = {
                "slide_title": slide_title,
                "research_results": slide_results
            }

    return completed_results


def merge_research_results_into_slides(slides_dict, research_results_dict):
    """
    slides_dict êµ¬ì¡°ì— research_results_dict ë°ì´í„°ë¥¼ ë³‘í•©
    slides_dict[slide_id]['research_results'] ë¡œ ì‚½ì…
    """
    for slide_id, data in research_results_dict.items():
        if slide_id in slides_dict:
            slides_dict[slide_id]['research_results'] = data.get('research_results', {})
    return slides_dict
````
